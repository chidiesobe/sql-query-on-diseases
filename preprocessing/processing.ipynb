```python
import pandas as pd
```

```python
data = pd.read_csv('dataset/hivaids.csv')
data.shape
```

```python
data.head(1)
```

```python
# Get list of columns 
column_names = data.columns.tolist()
column_names
```

```python
# Update column names
data.columns = data.columns.str.replace("Deaths - ", "d ")
data.columns = data.columns.str.split(' -').str[0]
data.columns = data.columns.str.replace("'", "")
data.columns = data.columns.str.replace(" ", "_")
data.columns = data.columns.str.lower()
data.columns = data.columns.str.replace("/", "_")
data.columns = data.columns.str.replace(",", "")
data.columns = data.columns.str.replace("-", "")
```

```python
# Get new list of columns 
column_names = data.columns.tolist()
column_names
```

```python
data.head(1)
```

```python
# Check if 'code' that would be primary key is empty and update it accordingly
data['code'] = data.apply(lambda row: row['entity'] if pd.isnull(row['code']) else row['code'], axis=1)
data['code'] = data['code'].apply(lambda x: x[:3].upper() if '(' not in x else x.split(' (')[1][:-1].upper())

# data.to_csv('check.csv')
```

```python
# Define cause of death categories for grouping
categories = {
    'location':[
    'entity','code',
    ],
    'Infectious Diseases': [
        'code', 'year',
        'd_meningitis', 'd_malaria', 'd_tuberculosis', 'd_acute_hepatitis', 'd_lower_respiratory_infections'
    ],
    'NonCommunicable Diseases': [
         'code', 'year',
        'd_alzheimers_disease_and_other_dementias', 'd_parkinsons_disease', 'd_cardiovascular_diseases',
        'd_neoplasms', 'd_diabetes_mellitus', 'd_chronic_kidney_disease', 'd_chronic_respiratory_diseases',
        'd_cirrhosis_and_other_chronic_liver_diseases', 'd_digestive_diseases'
    ],
    'External Causes': [
      'code', 'year',
        'd_drowning', 'd_interpersonal_violence', 'd_exposure_to_forces_of_nature', 'd_road_injuries',
        'd_poisonings', 'd_fire_heat_and_hot_substances'
    ],
    'Malnutrition and Maternal Health': [
      'code', 'year',
        'd_nutritional_deficiencies', 'd_proteinenergy_malnutrition', 'd_maternal_disorders'
    ],
    'Mental Health and Substance Abuse': [
     'code', 'year',
        'd_hiv_aids', 'd_drug_use_disorders', 'd_selfharm'
    ],
    'Other Causes': [
       'code', 'year',
        'd_alcohol_use_disorders', 'd_neonatal_disorders', 'd_conflict_and_terrorism',
        'd_diarrheal_diseases', 'd_environmental_heat_and_cold_exposure'
    ],
     'Years': [
       'year'
    ]
}

# Create CSV files for each category
for category, columns in categories.items():
    category_data = data[columns]
    category_data.to_csv(f'split_dataset/{category.replace(" ", "_").lower()}.csv', index=False)

print("CSV files created for each category.")
```
